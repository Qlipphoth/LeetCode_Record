## 45. 跳跃游戏 Ⅱ

```python
from typing import List, Optional
from collections import deque, defaultdict, Counter
```

```python
class Solution:
    def jump(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [(0, nums[0])]
        for i in range(1, n):
            step = float("inf")
            for x, y in dp:
                if y >= i: step = min(step, x + 1)
            dp.append((step, i + nums[i]))
        return dp[-1][0]
```

```python
nums = [2,3,0,1,4]
```

```python
s = Solution()
s.jump(nums)
```

```python
class Solution:
    def jump(self, nums: List[int]) -> int:
        far = step = end = 0
        for i in range(len(nums) - 1):
            far = max(far, i + nums[i])
            if i == end: end, step = far, step + 1
        return step
```

### 卡塔兰数

```python
class Solution:
    def numTrees(self, n: int) -> int:
        def dp(x):
            if x in memo: return memo[x]
            ans = 0
            if x & 1: ans += dp(x // 2) * dp(x // 2)
            for i in range(x // 2):
                ans += 2 * dp(i - 0) * dp(x - i - 1)
            memo[x] = ans
            return memo[x]
        
        memo = {0: 1, 1: 1}
        return dp(n) 
```

```python
s = Solution()
s.numTrees(19)
```

## 221. 最大正方形

$$dp(i, j) = min(dp(i - 1, j), dp(i, j - 1), dp(i - 1, j - 1)) + 1$$

```python
from typing import List
class Solution:
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        def dp(i, j):
            if i < 0 or j < 0 or matrix[i][j] == "0": return 0
            if (i, j) in memo: return memo[(i, j)]
            memo[(i, j)] = min(dp(i - 1, j), dp(i, j - 1), dp(i - 1, j - 1)) + 1
            return memo[(i, j)]
        
        ans = 0
        memo = {}
        m, n = len(matrix), len(matrix[0])
        for i in range(m):
            for j in range(n):
                ans = max(ans, dp(i, j))
        # print(memo)
        return ans
```
