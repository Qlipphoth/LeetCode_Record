```python
def constructTree(l:list) -> Optional[TreeNode]:
    head = TreeNode(val=l.pop(0))
    queue = [head]
    while l and queue:
        cur = queue.pop(0)
        lval = l.pop(0)
        if lval:
            curl = TreeNode(val=lval)
            queue.append(curl)
            cur.left = curl
        if l:
            rval = l.pop(0)
            if rval:
                curr = TreeNode(val=rval)
                queue.append(curr)
                cur.right =  curr
#         print(queue)
    return head
```

## 145. 二叉树的后序遍历

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        def DFS(node):
            if not node:
                return 
            DFS(node.left)
            DFS(node.right)
            res.append(node.val)
        return res
```

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        stk, res = [], []
        cur = root
        while stk or cur:
            while cur:
                res.append(cur.val)
                stk.append(cur)
                cur = cur.right
            cur = stk.pop().left
        return res[::-1]
```

后续： 左右中；相当于：中右左 的倒序，而前序遍历为： 中左右；故利用前序思路即可

```python
# 中序： 左自右
def inorderTraversal(root):
    stk, res = [], []
    cur = root
    while stk or cur:
        while cur:
            stk.append(cur)
            cur = cur.left
        cur = stk.pop()
        res.append(cur.val)
        cur = cur.right
    return res
```

```python
# 前序： 自左右
def preorderTraversal(root):
    stk, res = [], []
    cur = root
    while stk or cur:
        while cur:
            res.append(cur.val)
            stk.append(cur)
            cur = cur.left
        cur = stk.pop().right
    return res
```

```python
class Solution:
    def oneEditAway(self, first: str, second: str) -> bool:
        l1, l2 = len(first), len(second)
        if l1 == l2:
            cnt = 0
            for ch1, ch2 in zip(first, second):
                if ch1 != ch2: cnt += 1
            return True if cnt <= 1 else False
        if abs(l1 - l2) > 1: return False
        def compare(l, s):
            if s == '': return True
            idx1 = idx2 = 0
            while idx1 < len(l):
                if l[idx1] == s[idx2]:
                    idx2 += 1
                    if idx2 == len(s): return True
                idx1 += 1
            print(idx2)
            return False
        return compare(first, second) if l1 > l2 else compare(second, first)
```

```python
class Solution:
    def permutation(self, S: str) -> List[str]:
        if not S or len(S) <= 1: return []

        res, S, length = [], sorted(S), len(S)  # 排下序，使重复的都在一起

        def backtrack(used, paths):
            if length == len(paths):
                res.append(''.join(paths))
                return

            for i in range(length):
                if used[i]:
                    continue  # 已经选择过的不需要再放进去了
                if i > 0 and S[i] == S[i - 1] and not used[i - 1]:
                    continue  # 如果当前节点与他的前一个节点一样，并其他的前一个节点已经被遍历过了，那我们也就不需要了。

                used[i] = True
                paths.append(S[i])
                backtrack(used, paths)
                used[i] = False
                paths.pop()  # 还原used与paths的状态

        backtrack([False] * length, [])
        return res
```

## 523. 连续的子数组和

```python
from typing import List
from collections import deque
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        res = deque()
        for n in nums:
            res.append(n % k)
            if sum(res) > k:
                res.popleft()
            if sum(res) == k and len(res) > 1:
                return True
        return False
```

```python
nums = [23,2,4,6,6]
k = 7
# nums = [23,2,4,6,7], k = 6
# nums = [23,2,6,4,7], k = 6
s = Solution()
s.checkSubarraySum(nums, k)
```

满足条件的子数组可以由相应的前缀和相减得到，而满足子数组之和为k的倍数的条件为这两个前缀对k的取余相同（即相差为k的倍数），故要做的事情变为：枚举前缀和，将相应的余数保存，若遇到相同的余数则返回True，否则返回False

```python
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        presum = 0
        memo = set()
        for n in nums:
            last = presum
            presum += n
            presum %= k
            if presum in memo:
                return True
            memo.add(last)
            print(memo)
        return False
```

## 468. 验证IP地址

```python
class Solution:
    def validIPAddress(self, queryIP: str) -> str:
        ans1 = "Neither"
        if not queryIP: return ans1
        if "." in queryIP and ":" in queryIP: return ans1
        if "." in queryIP:
            res = queryIP.split(".")
            if len(res) != 4: return ans1
            for IP in res:
                L = len(IP)
                if not L: return ans1
                if IP[0] == "0" and L > 1: return ans1
                for ch in IP:
                    if not ch.isdigit(): return ans1
                if int(IP) > 255 or int(IP) < 0: return ans1
            return "IPv4"
        if ":" in queryIP:
            res = queryIP.split(":")
            if len(res) != 8: return ans1
            for IP in res:
                L = len(IP)
                if L > 4 or L < 1: return ans1
                for ch in IP:
                    if ch.islower() and ch > "f": return ans1
                    if ch.isupper() and ch > "F": return ans1
            return "IPv6"
```

## 33. 搜索旋转排序数组

```python
from typing import List
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        return nums.index(target) if target in nums else -1
```

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        idx = 0
        for i in range(1,len(nums)):
            if nums[i-1] > nums[i]:
                idx = i
                break
        res = [(0, idx-1),(idx, len(nums)-1)]
        for left, right in res:
            while left <= right:
                mid = (right + left)//2
                if nums[mid] < target:
                    left = mid + 1
                elif nums[mid] > target:
                    right = mid - 1
                else: return mid
        return -1
```

```python
nums = [4,5,6,7,0,1,2]
target = 2
s = Solution()
s.search(nums, target)
```

将数组一分为二，其中一定有一个是有序的，另一个可能是有序，也能是部分有序。
此时有序部分用二分法查找。无序部分再一分为二，其中一个一定有序，另一个可能有序，可能无序。就这样循环. 

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        if not nums:
            return -1
        l, r = 0, len(nums) - 1
        while l <= r:
            mid = (l + r) // 2
            if nums[mid] == target:
                return mid
            if nums[0] <= nums[mid]:
                if nums[0] <= target < nums[mid]:
                    r = mid - 1
                else:
                    l = mid + 1
            else:
                if nums[mid] < target <= nums[len(nums) - 1]:
                    l = mid + 1
                else:
                    r = mid - 1
        return -1
```

## 998. 最大二叉树 Ⅱ

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
from typing import Optional
class Solution:
    def insertIntoMaxTree(self, root, val: int) -> Optional[TreeNode]:
        father = left = None
        cur = root
        while cur:
            if cur.val > val:
                father = cur
                cur = cur.right
                left = cur
            else: break
        new = TreeNode(val)
        if father:
            father.right = new
            if left:
                new.left = left
            return root
        else: 
            new.left = root
            return new
```

沿着右子节点遍历，找到比val大的节点，作为父节点，该节点的右子树作为val节点的左子树

```python
def constructTree(l:list) -> Optional[TreeNode]:
    head = TreeNode(val=l.pop(0))
    queue = [head]
    while l and queue:
        cur = queue.pop(0)
        lval = l.pop(0)
        if lval:
            curl = TreeNode(val=lval)
            queue.append(curl)
            cur.left = curl
        if l:
            rval = l.pop(0)
            if rval:
                curr = TreeNode(val=rval)
                queue.append(curr)
                cur.right =  curr
#         print(queue)
    return head
```

```python
def inorderTraversal(root):
    res = []
    def Traversal(subroot):
        if not subroot:
            return 
        Traversal(subroot.left)
        res.append(subroot.val)
        Traversal(subroot.right)
    Traversal(root)
    return res
```

```python
from typing import Optional
class Solution:
    def insertIntoMaxTree(self, root, val: int) -> Optional[TreeNode]:
        last = None
        cur = root
        while cur:
            if cur.val < val: break
            last = cur
            cur = cur.right
        new = TreeNode(val)
        if last:
            if last.right:
                new.left = last.right
            last.right = new
            return root
        new.left = root
        return new
```

## 89. 格雷编码

```python
class Solution:
    def grayCode(self, n: int) -> List[int]:
        ans = [0]
        for i in range(1, n + 1):
            for j in range(len(ans) - 1, -1, -1):
                ans.append(ans[j] | (1 << (i - 1)))
        return ans
```

## 456. 132模式

错误的

```python
class Solution:
    def find132pattern(self, nums: List[int]) -> bool:
        stack = []
        k = -(10 ** 9 + 7)
        for i in range(len(nums) - 1,-1,-1):
            if nums[i] < k:
                return True
            while stack and stack[-1] < nums[i]:
                k = max(k,stack.pop())
            stack.append(nums[i])
        return False
```

```python
from sortedcotainers import SortedList
# 枚举3
class Solution:
    def find132pattern(self, nums: List[int]) -> bool:
        n = len(nums)
        if n < 3:
            return False
        
        # 左侧最小值
        left_min = nums[0]
        # 右侧所有元素
        right_all = SortedList(nums[2:])
        
        for j in range(1, n - 1):
            if left_min < nums[j]:
                index = right_all.bisect_right(left_min)
                if index < len(right_all) and right_all[index] < nums[j]:
                    return True
            left_min = min(left_min, nums[j])
            right_all.remove(nums[j + 1])

        return False
```
